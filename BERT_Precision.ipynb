```python
train2=train2[['comment_cleaned_spell','toxic']].rename(columns={'toxic':'labels','comment_cleaned_spell':'comment_cleaned'})
test2=test2[['comment_cleaned_spell','toxic']].rename(columns={'toxic':'labels','comment_cleaned_spell':'comment_cleaned'})
```

```python
dataset_train=Dataset.from_pandas(train2)
dataset_test=Dataset.from_pandas(test2)
```

```python
dataset_train
```

```python
from transformers import AutoTokenizer

tokenizer=AutoTokenizer.from_pretrained("bert-base-cased")

def tokenizer_function(data):
  return tokenizer(data['comment_cleaned'],padding='max_length',truncation=True)

tokenized_train=dataset_train.map(tokenizer_function)
tokenized_test=dataset_test.map(tokenizer_function)
```

```python
from transformers import AutoModelForSequenceClassification

model=AutoModelForSequenceClassification.from_pretrained("prajjwal1/bert-tiny",num_labels=2)
```

```python
model
```

```python
for name,param in model.named_parameters():
  print(name)
```

```python
for name,param in model.named_parameters():
  if name.startswith("bert.embeddings"):
    param.requires_grad=False
  if name.startswith("bert.encoder.layer.0"):
    param.requires_grad=False
  if name.startswith("bert.encoder.layer.1"):
    param.requires_grad=False
```

```python
from transformers import TrainingArguments

training_args= TrainingArguments(output_dir="test_trainer")
```

```python
import numpy as np
import evaluate

metric=evaluate.load("precision")
# metric=evaluate.load("recall")
```

```python
def compute_metrics(eval_pred):
  logits,labels=eval_pred
  predictions=np.argmax(logits,axis=-1)
  return metric.compute(predictions=predictions,references=labels)#,metric2.compute(predictions=predictions,references=labels)

from transformers import TrainingArguments, Trainer
training_args= TrainingArguments(output_dir="test_trainer",evaluation_strategy="epoch",num_train_epochs=4)
```

```python
tokenized_train
```

```python
trainer= Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
trainer.save_model('/content/drive/MyDrive/Team Blue NLP/Bert_precision')
```
